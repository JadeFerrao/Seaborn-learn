```python
#importimg libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

seaborn has built in data-sets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
# setting a theme for the plots, automatic grid setting
sns.set_theme()
```

```python
# visualize the data & differentiates automatically based on color, symbol unlike matplot
sns.relplot(data=tips, x ='total_bill',y='tip',col='time' ,hue='smoker' ,style='smoker', size='size')
```

using a matplotlib dataset, housing

```python
from sklearn.datasets import fetch_openml
import pandas as pd

# Load the California housing dataset
california_housing = fetch_openml(name="california_housing")

# Create a DataFrame from the dataset
house = pd.DataFrame(data=california_housing.data, columns=california_housing.feature_names)
house['PRICE'] = california_housing.target

#printing the table
house.head()
```

distribution plot

```python
sns.displot(house['PRICE'])
```

```python
sns.distplot(house['PRICE'])
```

Correlation:  
• Positive Correlation  
• Negative Correlation  

Heat map

```python
correlation = house.corr()
```

```python
# constructing a Heat Map
#lighter = negative value
#darker = positive value

plt.figure(figsize=(10,10))
sns.heatmap (correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size' :8}, cmap= 'Blues')
```
